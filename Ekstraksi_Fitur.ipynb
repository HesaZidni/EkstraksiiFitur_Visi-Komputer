```python
#upload file
from google.colab import files
file = files.upload()

#analisa gambar ke histogram
import cv2
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Load gambar ---
image_path = 'apple-3529407_1280.jpg'
image = cv2.imread(image_path)

# --- 2. Periksa apakah gambar berhasil dibaca ---
if image is None:
    print("Gambar tidak ditemukan.")
else:
    # --- 3. Konversi gambar ke RGB ---
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # --- 4. Tampilkan gambar asli ---
    plt.figure(figsize=(6, 6))
    plt.imshow(image_rgb)
    plt.title("Gambar Asli")
    plt.axis("off")
    plt.show()

    # --- 5. Hitung histogram warna ---
    colors = ('r', 'g', 'b')
    histogram_data = {}
    plt.figure(figsize=(10, 5))

    for i, color in enumerate(colors):
        hist = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
        histogram_data[color] = hist
        plt.plot(hist, color=color, label=f'{color.upper()} Channel')
        plt.xlim([0, 256])

    plt.title('Histogram Warna (RGB)')
    plt.xlabel('Intensitas Warna')
    plt.ylabel('Jumlah Piksel')
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- 6. Analisis warna dominan ---
    total_intensity = {color: np.sum(hist) for color, hist in histogram_data.items()}
    dominant_color = max(total_intensity, key=total_intensity.get)

    print("Total Intensitas Warna:")
    for color in colors:
        print(f"  {color.upper()}: {int(total_intensity[color])}")

    print(f"\n>> Warna dominan dalam gambar: {dominant_color.upper()}")


import cv2
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Load gambar ---
image_path = 'apple-3529407_1280.jpg'
image = cv2.imread(image_path)

# --- 2. Periksa apakah gambar berhasil dimuat ---
if image is None:
    print("Gambar tidak ditemukan.")
else:
    # --- 3. Konversi ke RGB dan HSV ---
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # --- 4. Tampilkan gambar RGB dan HSV ---
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title("Gambar RGB")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB))  # tampilkan HSV dalam warna RGB
    plt.title("Gambar HSV")
    plt.axis("off")
    plt.show()

    # --- 5. Histogram RGB ---
    colors_rgb = ('r', 'g', 'b')
    histogram_rgb = {}
    plt.figure(figsize=(10, 4))
    for i, color in enumerate(colors_rgb):
        hist = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
        histogram_rgb[color] = hist
        plt.plot(hist, color=color, label=f'{color.upper()} Channel')
    plt.title('Histogram RGB')
    plt.xlabel('Intensitas')
    plt.ylabel('Jumlah Piksel')
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- 6. Histogram HSV ---
    channels_hsv = ['Hue', 'Saturation', 'Value']
    colors_hsv = ['m', 'c', 'y']
    histogram_hsv = {}
    plt.figure(figsize=(10, 4))
    for i, channel in enumerate(channels_hsv):
        hist = cv2.calcHist([image_hsv], [i], None, [256], [0, 256])
        histogram_hsv[channel] = hist
        plt.plot(hist, color=colors_hsv[i], label=f'{channel}')
    plt.title('Histogram HSV')
    plt.xlabel('Intensitas')
    plt.ylabel('Jumlah Piksel')
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- 7. Analisis Dominasi Warna (RGB) ---
    total_intensity_rgb = {c: np.sum(histogram_rgb[c]) for c in colors_rgb}
    dominant_rgb = max(total_intensity_rgb, key=total_intensity_rgb.get)

    print(">> Total Intensitas Warna RGB:")
    for c in colors_rgb:
        print(f"   {c.upper()}: {int(total_intensity_rgb[c])}")
    print(f">> Warna Dominan RGB: {dominant_rgb.upper()}")

    # --- 8. Analisis Dominasi HSV (opsional) ---
    total_intensity_hsv = {ch: np.sum(histogram_hsv[ch]) for ch in channels_hsv}
    dominant_hsv = max(total_intensity_hsv, key=total_intensity_hsv.get)

    print("\n>> Total Intensitas Warna HSV:")
    for ch in channels_hsv:
        print(f"   {ch}: {int(total_intensity_hsv[ch])}")
    print(f">> Komponen HSV Dominan: {dominant_hsv}")

# --- 9. Analisis Kematangan Berdasarkan Warna (HSV) ---

# Buat masker HSV untuk warna merah matang
# OpenCV HSV: H ∈ [0,179], S ∈ [0,255], V ∈ [0,255]
# Warna merah menyebar di dua rentang hue: 0–10 dan 160–180

# Rentang bawah merah
lower_green1 = np.array([0, 100, 100])
upper_green1 = np.array([10, 255, 255])

# Rentang atas merah
lower_green2 = np.array([160, 100, 100])
upper_green2 = np.array([179, 255, 255])

# Buat dua masker
mask1 = cv2.inRange(image_hsv, lower_green1, upper_green1)
mask2 = cv2.inRange(image_hsv, lower_green2, upper_green2)

# Gabungkan kedua masker
mask_red = cv2.bitwise_or(mask1, mask2)

# Hitung persentase area yang termasuk warna matang
total_pixels = image_hsv.shape[0] * image_hsv.shape[1]
red_pixels = cv2.countNonZero(mask_red)
maturity_percent = (red_pixels / total_pixels) * 100

# Tampilkan masker (opsional)
plt.figure(figsize=(6, 6))
plt.imshow(mask_red, cmap='gray')
plt.title('Masker Warna Hijau (BELUM MATANG)')
plt.axis('off')
plt.show()

# --- 10. Cetak hasil analisis kematangan ---
print(f">> Estimasi Area Buah Matang: {maturity_percent:.2f}% dari total gambar.")

if maturity_percent > 50:
    print(">> Buah kemungkinan besar SUDAH MATANG.")
elif maturity_percent > 20:
    print(">> Buah kemungkinan dalam TAHAP MATANG.")
else:
    print(">> Buah kemungkinan BELUM MATANG.")
```
